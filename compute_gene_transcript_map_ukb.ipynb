```python
import gzip
import pandas as pd

chromosomes = [str(i) for i in range(1, 23)] + ['X', 'Y', 'MT']

combined_df = pd.DataFrame()

for chrom in chromosomes:
    fname = f'chr{chrom}.filtered.vep.tsv.gz'
    try:
        df = pd.read_csv(fname, sep='\t', compression='gzip')
        df = df[['CHROM', 'ENST', 'ENSG', 'SYMBOL']].drop_duplicates()
        combined_df = pd.concat([combined_df, df], ignore_index=True)
        print(f"{fname}: loaded {len(df)} unique rows.")
    except FileNotFoundError:
        print(f"File not found: {fname}, skipping.")
    except Exception as e:
        print(f"Error reading {fname}: {e}")

# Optionally drop duplicates across all chromosomes
combined_df = combined_df.drop_duplicates()

print(f"Combined total: {len(combined_df)} unique rows.")
# To save:
# combined_df.to_csv('combined_genes.tsv', sep='\t', index=False)
```
